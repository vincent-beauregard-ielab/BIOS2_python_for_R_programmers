## Matplotlib
* Python's plotting library. Similar in usage to R base plotting functions.
* Many customizations possible for appearance and layout.
* Many types of plots available (line, scatter, bar, histogram, boxplot, ...).
* See [matplotlib gallery](https://matplotlib.org/gallery/index.html) for examples and [documentation](https://matplotlib.org/contents.html) for details.

### Demo

```python
import matplotlib.pyplot as plt
import numpy as np

# Lets create a simple line plot using the plot function.
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# Let's create a simple plot using two random variables along the x axes.
x = np.arange(0, 1, 0.01)
y = np.random.randn(len(x) * 2).reshape(2, len(x))

# Add a bias to the second line.
y[1] += 1

# We can plot the two lines using the plot function.
plt.scatter(x, y[0], label='Observed variable 1')
plt.scatter(x, y[1], label='Observed variable 2')

# Add title, axis labels and legend.
plt.title('Two random variables')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Let's create an histogram comparing the two random variables.
plt.hist(y[0], label='Observed variable 1', alpha=0.5) # alpha is the transparency
plt.hist(y[1], label='Observed variable 2', alpha=0.5)

plt.title('Histogram of two random variables')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Use the subplot function to create a figure with two subplots with the scatter plots and histograms.
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].scatter(x, y[0], label='Observed variable 1')
ax[0].scatter(x, y[1], label='Observed variable 2')
ax[0].set_title('Scatter plot of two random variables')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend()

ax[1].hist(y[0], label='Observed variable 1', alpha=0.5)
ax[1].hist(y[1], label='Observed variable 2', alpha=0.5)
ax[1].set_title('Histogram of two random variables')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].legend()
plt.show()
```
